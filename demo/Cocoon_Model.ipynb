```python
#@title  Download example table (skip this step if you have your own table)
import requests

files = {
"completeddisposition.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completeddisposition.csv",
"completedtrans.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completedtrans.csv",
"completeddistrict.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completeddistrict.csv",
"CRM_Reviews.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/CRM_Reviews.csv",
"completedcard.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completedcard.csv",
"completedorder.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completedorder.csv",
"completedacct.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completedacct.csv",
"completedclient.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completedclient.csv",
"completedloan.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/completedloan.csv",
"LuxuryLoanPortfolio.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/LuxuryLoanPortfolio.csv",
"CRM_Events.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/CRM_Events.csv",
"CRM_Call_Center_Logs.csv": "https://raw.githubusercontent.com/Cocoon-Data-Transformation/cocoon/main/files/CRM_Call_Center_Logs.csv"
}

# Loop through the files dictionary
for file_name, url in files.items():
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Open file in binary write mode
        with open(file_name, "wb") as file:
            file.write(response.content)
        print(f"{file_name} downloaded successfully.")
    else:
        print(f"Failed to download {file_name}.")
```

```python
# @title Provide Data Warehouse Con

# In-mem duckdb loads the df
con = duckdb.connect(database=':memory:')
for csv_file in files:
    file_path = f'./{csv_file}'

    df = pd.read_csv(file_path)
    table_name = os.path.basename(file_path).split('.')[0]
    table_name = clean_table_name(table_name)

    query_widget = QueryWidget(con)
    con.register(table_name, df)

# # Snowflake: specify the con info and table ame
# con = snowflake.connector.connect(
#     # check out your url: {account}.snowflakecomputing.com
#     account="",
#     user="",
#     password="",
#     warehouse="",
#     database="",
#     schema="",
# )
# # Please enter the table to stage/clean
# # Make sure it's a table (queries over view can be slow)
```
