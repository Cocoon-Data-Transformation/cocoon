```python
file_path = "./files/patients.csv"

# create a pandas dataframe from the csv file
df = pd.read_csv(file_path)

# pass the dataframe and the table name to the cocoon documenter
cocoon_documenter = DocumentedData(df, table_name="patients")

# optionally, you can load past documentat
# cocoon_documenter.read_document_from_disk('./files/patients_docu.json')
```

```python
cocoon_documenter.start_document()
```

```python
# remember to save the document to disk
cocoon_documenter.write_document_to_disk('./files/patients_docu.json')
```

```python
# display the document
cocoon_documenter
```

```python
cocoon_documenter.save_as_html('./files/patients.html')
```

```python
cocoon_cleaner = DataCleaning(cocoon_documenter)
```

```python
cocoon_cleaner
```

```python
cleaned_df = cocoon_cleaner.run_codes()

cleaned_df
```

```python
cocoon_cleaner.print_codes()
```

```python
# pass the documenter to the cocoon transformer
cocoon_transformer = CDMTransformation(cocoon_documenter)

# optionally, you can load past transformation
# cocoon_transformer.read_document_from_disk('./files/patients_transformer.json')
```

```python
cocoon_transformer.display_database()
```

```python
cocoon_transformer.start()
```

```python
cocoon_transformer.write_document_to_disk('./files/patients_transformer.json')
```

```python
cocoon_transformer.pipeline.print_codes()
```
